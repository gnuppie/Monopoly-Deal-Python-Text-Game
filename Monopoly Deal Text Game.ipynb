```python
#Shuffle deck function, and draw card/discard card functions.

# Drawing like irl, from top of a shuffled deck. But need to add discard pile back and reshuffle later.
#from random import shuffle

def check_shuffle_deck(deck,discard_pile):
    if len(deck) < 5:
        print("Deck has less than 5 cards remaining, adding cards from discard pile and reshuffling...")
        sleep(1)
        deck.extend(discard_pile)
        discard_pile.clear()
        seed(randint(0,1000))
        shuffle(deck)

        
def draw_card(deck,player_hand):
    drawn = deck[0]
    player_hand.append(drawn)
    deck.remove(drawn)
    return None
# Draws one random card from the first variable deck, and then removes it from deck and appends it to the player's hand

def discard_card(deck,player_hand,discard_pile):
    print(player_hand)
    discard = input("Which card do you want to discard [Please write the EXACTLY identical name of the card]: ")
    if discard not in player_hand:
        print("You did not input an eligible card. Try again.")
        discard_card(deck,player_hand,discard_pile)
    else:
        discard_pile.append(discard)
        player_hand.remove(discard)
    return None
# Allows the player to pick which card to discard from the hand by inputting the name of the card, then removes it from hand
# and appends it to the deck


#Test Case
#print(deck)
#shuffle(deck)
#print(deck)
#player_hand = []
#draw_card(deck,player_hand)
#print(player_hand)
#print(deck)
#discard_card(deck,player_hand,discard_pile)
#print(player_hand)
#print(deck)
#print(discard_pile)

#Reset the deck and discard pile
#deck = complete_deck[:]
#discard_pile = []
```

```python
#Start_turn function

def start_turn(n):
#This is to set the variable names that would be used in the functions used during the turn, for ease of coding
    if n == 1:
        player = player1
        player_hand = player1["Hand"]
        player_property = player1["Property"]
        player_set = player1["Set"]
        player_name = player1["Name"]
    elif n == 2:
        player = player2
        player_hand = player2["Hand"]
        player_property = player2["Property"]
        player_set = player2["Set"]
        player_name = player2["Name"]
    elif n == 3:
        player = player3
        player_hand = player3["Hand"]
        player_property = player3["Property"]
        player_set = player3["Set"]
        player_name = player3["Name"]
    elif n == 4:
        player = player4
        player_hand = player4["Hand"]
        player_property = player4["Property"]
        player_set = player4["Set"]
        player_name = player4["Name"]
    else:
        return "Error, player number is not within range of 1-4."


    return player,player_hand,player_property,player_set,player_name
```

```python
#Misc. functions that are to be used in the various action card functions

#straightforward functions
def take_property(taker_property,target_property,card_taken):
    taker_property.append(card_taken)
    target_property.remove(card_taken)
    
def trade_property(taker_property, target_property, card_taken, card_given):
    taker_property.append(card_taken)
    target_property.remove(card_taken)
    target_property.append(card_given)
    taker_property.remove(card_given)
    
def take_set(taker_set,target_set,set_taken):
    taker_set.append(set_taken)
    target_set.remove(set_taken)

#function for those action cards that require a target to be chosen
def choose_target(player_number):
    player_names = []
    for individual in total_players:
        if individual["Number"] != player_number:
            player_names.append(individual["Name"])
        else:
            your_name = individual["Name"]
    print(player_names)
    target = input("Who is your target? [Cannot pick yourself ({}) [Please write the EXACT identical name]]: ".format(your_name))
    if target == your_name:
        print("You cannot pick yourself. Choose your target again.")
        choose_target(player_number)
    elif target not in player_names:
        print("Invalid target. Choose your target again.")
        choose_target(player_number)
    else:
#target variables are set to the player chosen
        for individual in total_players:
            if individual["Name"] == target:
                target_hand = individual["Hand"]
                target_property = individual["Property"]
                target_set = individual["Set"]
    
    return target_hand, target_property, target_set
```

```python
#End turn functions

# This function checks whether the player's hand
def check_hand_no(player_hand):
    if len(player_hand) == 0:
        print("You have zero cards in your hand, drawing 5 cards from deck...")
        while len(player_hand) < 5:
            draw_card(deck,player_hand)
# Draws 5 cards if the player has no more cards in his hand.
    elif len(player_hand) > 7:
        while len(player_hand) > 7:
            discard_card(deck,player_hand,discard_pile)
# Discards all cards until the player has a hand of only 7 cards maximum
    print("Your hand is now as follows: {}".format(player_hand))
    return None

# Test case 1 (9 cards in hand) - if using test case, please reset deck after to original.
#player_hand = ['Brown_1','Brown_2','Brown_3','Blue_1','Blue_2','Blue_3','Green_1','Green_2','Green_3']
#check_hand_no(player_hand)
#print(discard_pile)
#deck = complete_deck[:]

# Test case 2 (0 cards in hand) - if using test case, please reset deck after to original.
#player_hand = []
#check_hand_no(player_hand)
#deck = complete_deck[:]

#-------------------------------------------------------------------------------------------------------------------------
#function to check if there are any complete sets within the player's property
def check_complete(player_property,player_set):
    brown = 0
    blue = 0
    green = 0
    orange = 0
    pink = 0
    red = 0
    yellow = 0
    black = 0
    white = 0
    cyan = 0
# we set counters for each colour to count the amount of properties of each colour within the player_property
    for land in player_property:
        if land in complete_brown:
            brown += 1
        elif land in complete_blue:
            blue += 1
        elif land in complete_green:
            green += 1
        elif land in complete_orange:
            orange += 1
        elif land in complete_pink:
            pink += 1
        elif land in complete_red:
            red += 1
        elif land in complete_yellow:
            yellow += 1
        elif land in complete_black:
            black += 1
        elif land in complete_white:
            white += 1
        elif land in complete_cyan:
            cyan += 1

# checks which properties have complete sets. NO ELIFS are used as it would only cause 1 iteration, when we want to check
# if there are multiple instances of completed sets on the same turn that we are checking
    if brown == 3:
        for i in complete_brown:
              player_property.remove(i)
        player_set.append('Brown')

    if blue == 3:
        for i in complete_blue:
            player_property.remove(i)
        player_set.append('Blue')                

    if green == 3:
        for i in complete_green:
            player_property.remove(i)
        player_set.append('Green')
        
    if orange == 3:
        for i in complete_orange:
              player_property.remove(i)
        player_set.append('Orange')        
        
    if pink == 3:
        for i in complete_pink:
            player_property.remove(i)
        player_set.append('Pink') 
        
    if red == 3:
        for i in complete_red:
            player_property.remove(i)
        player_set.append('Red')  
        
    if yellow == 3:
        for i in complete_yellow:
            player_property.remove(i)
        player_set.append('Yellow')
        
    if black == 3:
        for i in complete_black:
            player_property.remove(i)
        player_set.append('Black')     
        
    if white == 3:
        for i in complete_white:
            player_property.remove(i)
        player_set.append('White')  
        
    if cyan == 3:
        for i in complete_cyan:
             player_property.remove(i)
        player_set.append('Cyan')
    
    print("These are your uncompleted properties: {},\nand these are your completed sets: {}".format(player_property,player_set))

# Test case
#player_property = ['Brown_1','Brown_2','Brown_3','Blue_1','Blue_2','Blue_3','Green_1','Green_2']
#player_set = []
#check_complete(player_property,player_set)

#-------------------------------------------------------------------------------------------------------------------------
#Function to check if the player has won
def check_win(player_set,player_name):
    if len(player_set) >= 3:
        return player_name
    else:
        return None

#-------------------------------------------------------------------------------------------------------------------------
#OVERALL end_turn() function
def end_turn(player_hand,player_property,player_set,player_name):
    check_complete(player_property,player_set)
    winner = check_win(player_set,player_name)
    if winner == None:
        check_hand_no(player_hand)
        return winner
    else:
        return winner
```

```python
#Change turn function

#For clearing the output for the next user turn
#from IPython.display import clear_output

#Changes the player number to the next player
def change_turn(player_number,player_name):
    clear_output(wait=True)
    print("Ending {}'s turn...".format(player_name))
    if player_number < len(player_ls):
        player_number += 1
    elif player_number == len(player_ls):
        player_number = 1
    
    return player_number
```

```python
#Master Function

def monopoly_deal():
    print("MONOPOLY DEAL PYTHON TEXT GAME\nProgrammed by gnuppie\nPlease note that this version still contains some bugs that could potentially crash the game.")
    print("---------------------------------------------------------------------------------------")
    input("Press enter to start game.")
#Shuffle the sample deck with a randomised seed number between 0 to 1000
    print("Shuffling the deck...")
    sleep(1)
    seed(randint(0,1000))
    shuffle(complete_deck)
#Reset for every start of game
    global deck
    deck = complete_deck[:]
    global discard_pile
    discard_pile = []
    total_player_number = intro()
    start_game(total_player_number)

    winner = None
    global player_number
    player_number = 1
    while winner is None:
        player,player_hand,player_property,player_set,player_name = start_turn(player_number)
        print("{} is drawing two cards...".format(player_name))
        draw_card(deck,player_hand)
        draw_card(deck,player_hand)
        sleep(1)
        action(player_hand,player_property, player_set,player_name)
        winner = end_turn(player_hand,player_property,player_set,player_name)
        sleep(3)
        player_number = change_turn(player_number,player_name)
    
    return "The minimum of 3 complete sets has been achieved. The winner is {}!".format(winner)
```
